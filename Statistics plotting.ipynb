```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
LOG_DIR = "logs"
filename= "log-1527118578.9081967.txt"
path = os.path.join(LOG_DIR, filename)
data = pd.read_csv(path,skipinitialspace=True)
```

```python

plt.figure(figsize=(20,10))
plt.title("Loss")
plt.ylim([0.5,0.74])
plt.plot(data.reconstruction_loss_train, '--o', label='Test reconstruction loss')
plt.plot(data['reconstruction_loss_test '], '--o', label='Train reconstruction loss')
plt.legend()
```

```python

plt.figure(figsize=(20,10))
plt.title("Loss")
plt.ylim([0,0.05])
plt.plot(data.test_loss, '--o', label='Test loss')
plt.plot(data.train_loss, '--o', label='Train loss')
plt.legend()
```

```python
plt.figure(figsize=(20,10))
plt.ylim([99,100])
plt.plot(data.test_accuracy, '--o')
```
